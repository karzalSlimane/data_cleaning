```python
#tache 3

import pandas as pd
import numpy as np
from scipy.stats import zscore
from sklearn.preprocessing import StandardScaler, MinMaxScaler

# Détection des valeurs aberrantes avec le score Z
def detect_outliers_z_score(data, threshold=3):
    z_scores = zscore(data.select_dtypes(include=[np.number]))
    abs_z_scores = np.abs(z_scores)
    outliers = (abs_z_scores > threshold)
    print("Valeurs aberrantes détectées avec le score Z :")
    return data[outliers.any(axis=1)]

# Détection des valeurs aberrantes avec la méthode IQR
def detect_outliers_iqr(data, column):
    Q1 = data[column].quantile(0.25)
    Q3 = data[column].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    outliers = data[(data[column] < lower_bound) | (data[column] > upper_bound)]
    print(f"\nValeurs aberrantes dans la colonne '{column}' avec la méthode IQR :")
    return outliers

# 1. Exemple d'utilisation du score Z pour détecter les valeurs aberrantes
outliers_z_score = detect_outliers_z_score(data_frame)
print(outliers_z_score)

# 2. Exemple d'utilisation de la méthode IQR pour détecter les valeurs aberrantes dans chaque colonne
for column in data_frame.select_dtypes(include=[np.number]).columns:
    outliers_iqr = detect_outliers_iqr(data_frame, column)
    print(outliers_iqr)

# Standardisation des données (score Z)
scaler = StandardScaler()
data_standardized = pd.DataFrame(scaler.fit_transform(data_frame.select_dtypes(include=[np.number])),
                                 columns=data_frame.select_dtypes(include=[np.number]).columns)
print("\nDonnées standardisées :")
data_standardized
```
